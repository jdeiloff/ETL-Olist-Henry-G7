## Conexión a la DB y generación de dataframes

```python
engine = create_engine(<access>)
customers = pd.read_sql_table('customers', engine)
geolocation=pd.read_sql_table('geolocation', engine)
order_items=pd.read_sql_table('order_items', engine)
order_payments=pd.read_sql_table('order_payments', engine)
order_reviews=pd.read_sql_table('order_reviews', engine)
orders=pd.read_sql_table('orders', engine)
products=pd.read_sql_table('products', engine)
sellers=pd.read_sql_table('sellers', engine)
product_category_name_translation=pd.read_sql_table('product_category_name_translation', engine)
calendar=pd.read_sql_table('calendar',engine)
```

## Uniones de dataframes

```python
df_exp=pd.merge(orders, order_reviews , on='order_id', how='inner')
df_exp=pd.merge(df_exp, customers , on='customer_id', how='inner')
df_exp=pd.merge(df_exp,order_items,on='order_id',how='inner')
df_exp=pd.merge(df_exp,products,on='product_id',how='inner')
df_exp=pd.merge(df_exp,geolocation,on='zip_code_prefix',how='inner')
```

```python
feature_group_active = folium.FeatureGroup(name='Active')
feature_group_unactive = folium.FeatureGroup(name='Unactive')
```

```python
map_clusters = folium.Map(location=[-15.89, -47.9], zoom_start=11)
# add markers to the map
marker_cluster_active = MarkerCluster()
marker_cluster_unactive =MarkerCluster()
for i in range (len(df_exp['latitude'])):
    if(df_exp["review_score"][i]>3):
        marker_active = folium.Marker(location=[df_exp['latitude'][i],df_exp['longitude'][i]],popup=df_exp['product_category_name'][i],icon = folium.Icon(color='green',icon='ok-sign'))
        marker_cluster_active.add_child(marker_active)
    else:
        marker_unactive = folium.Marker(location=[df_exp['latitude'][i],df_exp['longitude'][i]],popup=df_exp['product_category_name'][i],icon = folium.Icon(color='red',icon='exclamation-sign'))
        marker_cluster_unactive.add_child(marker_unactive)
```

```python
feature_group_active.add_child(marker_cluster_active)
feature_group_unactive.add_child(marker_cluster_unactive)
```
