## Creación de vistas temporales en Spark para conectarse a los RDDs por medio de Hive SQL

```python
%scala
customers.createOrReplaceTempView("customer")
orders.createOrReplaceTempView("order")
products.createOrReplaceTempView("product")
sellers.createOrReplaceTempView("seller")
order_items.createOrReplaceTempView("order_item")
order_payments.createOrReplaceTempView("order_payment")
order_reviews.createOrReplaceTempView("order_reviews")
geolocation.createOrReplaceTempView("geolocations")
product_category_name.createOrReplaceTempView("product_category_names")
calendar.createOrReplaceTempView("calendar")
```

## Querys de Hive SQL para generar las tablas Delta del DataWarehouse en Delta Lake

```python
%sql
CREATE TABLE customers USING DELTA AS
SELECT * FROM customer;
CREATE TABLE orders USING DELTA AS
SELECT * FROM order;
CREATE TABLE products USING DELTA AS
SELECT * FROM product;
CREATE TABLE sellers USING DELTA AS
SELECT * FROM seller;
CREATE TABLE order_items USING DELTA AS
SELECT * FROM order_item;
CREATE TABLE order_payments USING DELTA AS
SELECT * FROM order_payment;
CREATE TABLE order_reviews USING DELTA AS
SELECT * FROM order_reviews;
CREATE TABLE geolocation USING DELTA AS
SELECT * FROM geolocations;
CREATE TABLE product_category_name USING DELTA AS
SELECT * FROM product_category_names;
CREATE TABLE calendar USING DELTA AS
SELECT * FROM calendar;
```
