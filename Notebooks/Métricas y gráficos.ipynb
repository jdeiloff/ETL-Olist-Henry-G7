## Puntajes bajos por región

```python
region_entregas =pd.DataFrame()
region_entregas['totales']=df_exp['region'].value_counts()
region_entregas['Puntajes bajos'] = df_bajo['region'].value_counts()
region_entregas['Porcentaje de PB']=region_entregas['Puntajes bajos']*100/region_entregas['totales']
region_entregas['Puntajes bajos por entrega'] = entrega_baja['region'].value_counts()
region_entregas['Porcentajes de PN por entrega']=region_entregas['Puntajes bajos por entrega']*100/region_entregas['Puntajes bajos']
```

```python
plt.style.use('ggplot')
region_entregas[['Porcentaje de PB','Porcentajes de PN por entrega']].plot(kind ='bar')
plt.ylabel('Porcentaje')
plt.xlabel('Región')
plt.title('Porcentaje de puntuaciones bajas por región')
plt.show()
```

```python
colores= ['red','orange','yellow','green','blue']
region_entregas['Porcentaje de PB'].plot(kind='barh',color=colores)
plt.ylabel('Regiones')
plt.xlabel('Porcentaje de puntajes bajos')
plt.title('Puntuaciones bajas por region')
plt.show()
```

```python
#Clasificación categórica según n° de fotos disponibles por producto ordenado.

products.info()
print(products['photos_quantity'].unique())

products['photos_quantity']=products['photos_quantity'].fillna(0)

products['fotos_incluidas'] = products['photos_quantity'].apply(lambda x : 'Sin foto' if x == 0 else( 'Foto unica' if x == 1  else('Varias fotos' if x <= 3  else 'Muchas fotos')))

data_fotos= pd.merge(products,order_items,on='product_id',how='inner')
data_fotos.info()
#Borrado de columnas irrelevantes
data_fotos.drop(columns=['seller_id','shipping_limit_date','name_length','description_length','weight_g','length_cm','height_cm','width_cm','freight_value'],inplace=True)
```
